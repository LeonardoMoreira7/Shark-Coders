```python
from tkinter import *
from tkinter import messagebox

def calcular():
    try:
        peso = float(input_peso.get())
        altura = float(input_altura.get()) / 100 # Converter altura de cm para metros
        imc = peso / (altura ** 2)
        lb_resultado.config(text=f"IMC: {imc:.2f}") # .2f = reduzir as casas decimais para duas


        if imc < 18.5:
            imagem_path = "fotos/imc_magro.png"
        elif imc < 24.9:
            imagem_path = "fotos/imc_normal.png"
        elif imc < 29.9:
            imagem_path = "fotos/imc_sobrepeso.png"
        elif imc < 39.9:
            imagem_path = "fotos/imc_obeso.png"
        else:
            imagem_path = "fotos/imc_obeso_2.png"

        # Mostrar a imagem correspondente
        imagem = PhotoImage(file=imagem_path)
        lb_imagem.config(image=imagem)
        lb_imagem.image = imagem

    except ValueError:
        messagebox.showerror("Erro", "Por favor insira valores válidos para peso e altura.")

preto = '#000000'

root = Tk()
root.title("Agenda")
root.geometry("380x500+500+100")
root.wm_resizable(width=False, height=False)

lb_calculadora = Label(root, text="Calculadora de IMC", font="Time 20", fg=preto)
lb_calculadora.place(width=380, height=50, x=0, y=0)

lb_peso = Label(root, text="Peso(kg)", font="Time 10", anchor= "w")
lb_peso.place(width=60, height=20, x=10, y=80)
input_peso = Entry(root, font="Time 10")
input_peso.place(width=250, height=20, x=100, y=80)

lb_altura = Label(root, text="Altura(cm)", font="Time 10", anchor= "w")
lb_altura.place(width=65, height=20, x=10, y=120)
input_altura = Entry(root, font="Time 10")
input_altura.place(width=250, height=20, x=100, y=120)

b_calcular = Button(root, text="Calcular IMC", font="Time 10", command=calcular)
b_calcular.place(width=80, height=30, x=150, y=170)

lb_resultado = Label(root, text="", font="Arial 14")
lb_resultado.place(width=100, height=20, x=140, y=210)

lb_imagem = Label(root)
lb_imagem.place(width=80, height=200, x=150, y=250)




root.mainloop()

```

```python
from tkinter import *
from tkinter import messagebox

preto = '#000000'

root = Tk()
root.title("Agenda")
root.geometry("380x500+500+100")
root.wm_resizable(width=False, height=False)

lb_calculadora = Label(root, text="Calculadora de IMC", font="Time 20", fg=preto)
lb_calculadora.place(width=380, height=50, x=0, y=0)

lb_peso = Label(root, text="Peso(kg)", font="Time 10", anchor= "w")
lb_peso.place(width=60, height=20, x=10, y=80)
input_peso = Entry(root, font="Time 10")
input_peso.place(width=250, height=20, x=100, y=80)

lb_altura = Label(root, text="altura(cm)", font="Time 10", anchor= "w")
lb_altura.place(width=60, height=20, x=10, y=120)
input_altura = Entry(root, font="Time 10")
input_altura.place(width=250, height=20, x=100, y=120)

b_calcular = Button(root, text="Calcular IMC", font="Time 10")
b_calcular.place(width=80, height=30, x=150, y=170)

root.mainloop()

```

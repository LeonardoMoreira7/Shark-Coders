```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

def converter():
    origem = b_usd.get()
    para = b_brl.get()
    valor = float(input_valor.get())

    cotacao = requests.get("https://economia.awesomeapi.com.br/json/last/USD-BRL")
    cotacao = cotacao.json()
    cotacao_moeda = float(cotacao[origem+para]["bid"])
    resultado = round(valor * cotacao_moeda, 2)
    lb_valorconvertido.config(text=resultado)

preto = '#000000'
branco = '#ffffff'

root = Tk()
root.title("Conversor")
root.geometry("380x500+500+100")
root.wm_resizable(width=False, height=False)

lista = ["EUR", "USD", "BRL"]

lb_conversor = Label(root, text="Conversor de Moeda", font="Time 20", bg=preto, fg=branco)
lb_conversor.place(width=380, height=50, x=0, y=0)

lb_valorconvertido = Label(root, text="", font="Time 20", fg=preto)
lb_valorconvertido.place(width=380, height=80, x=0, y=100)

lb_de = Label(root, text="De", font="Time 11", fg=preto)
lb_de.place(width=50, height=50, x=70, y=140)

lb_para = Label(root, text="Para", font="Time 11", fg=preto)
lb_para.place(width=50, height=50, x=250, y=140)

b_usd = Button(root, text="USD", font="Time 15", fg=preto)
b_usd.place(width=50, height=30, x=70, y=180)

b_brl = Button(root, text="BRL", font="Time 15", fg=preto)
b_brl.place(width=50, height=30, x=250, y=180)

input_valor = Entry(root, font="Time 15")
input_valor.place(width=300, height=30, x=40, y=240)

b_converter = Button(root, text="Converter", font="Time 15", bg=preto, fg=branco, command=converter)
b_converter.place(width=250, height=30, x=65, y=320)


root.mainloop()

```

```python
import tkinter as tk
from tkinter import messagebox
import requests  # type: ignore

BASE_URL = "https://pokeapi.co/api/v2/pokemon/"

class PokemonApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Informações sobre Pokemons")
        self.root.geometry("400x300")
        self.root.configure(bg="#2E2E2E")

        self.digitar_pokemon_label = tk.Label(root, text="Digite o nome do Pokemon:", bg="#2E2E2E", fg="white")
        self.digitar_pokemon_label.pack(pady=10)

        self.digitar_pokemon_entry = tk.Entry(root, bg="#4E4E4E", fg="white")
        self.digitar_pokemon_entry.pack(pady=5)

        self.tipo_label = tk.Label(root, text="Tipo:", bg="#2E2E2E", fg="white")
        self.tipo_label.pack(pady=5)

        self.ps_label = tk.Label(root, text="PS:", bg="#2E2E2E", fg="white")
        self.ps_label.pack(pady=5)

        self.peso_label = tk.Label(root, text="Peso (kg):", bg="#2E2E2E", fg="white")
        self.peso_label.pack(pady=5)

        self.buscar_button = tk.Button(root, text="Buscar", command=self.buscar_pokemon, bg="#4E4E4E", fg="white")
        self.buscar_button.pack(pady=10)

        self.resultado_label = tk.Label(root, text="", bg="#2E2E2E", fg="white")
        self.resultado_label.pack(pady=10)

    def buscar_pokemon(self):
        url = BASE_URL + self.digitar_pokemon_entry.get().lower()
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            tipo = data["types"][0]["type"]["name"]
            ps = data["stats"][0]["base_stat"]
            peso = data["weight"]
            self.tipo_label.config(text=f"Tipo: {tipo}")
            self.ps_label.config(text=f"PS: {ps}")
            self.peso_label.config(text=f"Peso: {peso}")
            return tipo, ps, peso
        else:
            messagebox.showerror("Erro", "Pokemon não encontrado!")

def main():
    root = tk.Tk()
    app = PokemonApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

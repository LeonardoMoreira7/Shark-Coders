```python
from tkinter import *
from tkinter import messagebox

#--------------------------------------------------------------------------------------------------------------------
# cores

branco = '#ffffff'
azul = '#364a85'
vermelho = '#b53128'
amarelo = '#ffef08'

# Tela
root = Tk()
root.title("Agenda")
root.geometry("380x500+500+100")
root.wm_resizable(width=False, height=False)

# Labels e Entrys

lb_agenda = Label(root, text="Sharkcoders Python", font="Time 20", bg=amarelo, fg=vermelho, anchor="w", padx=10)
lb_agenda.place(width=380, height=50, x=0, y=0)

lb_nome = Label(root, text="Nome", font="Time 10", anchor= "w")
lb_nome.place(width=60, height=20, x=10, y=70)
input_nome = Entry(root, font="Time 10")
input_nome.place(width=250, height=20, x=100, y=70)

lb_tel = Label(root, text="Telemóvel", font="Time 10", anchor= "w")
lb_tel.place(width=60, height=20, x=10, y=110)
input_tel = Entry(root, font="Time 10")
input_tel.place(width=250, height=20, x=100, y=110)

lb_end = Label(root, text="Endereço", font="Time 10", anchor= "w")
lb_end.place(width=60, height=20, x=10, y=170)
input_end = Entry(root, font="Time 10")
input_end.place(width=250, height=20, x=100, y=170)

lb_dist = Label(root, text="Distrito", font="Time 10", anchor= "w")
lb_dist.place(width=60, height=20, x=10, y=210)
input_dist = Entry(root, font="Time 10")
input_dist.place(width=80, height=20, x=100, y=210)

lb_pais = Label(root, text="País", font="Time 10", anchor= "w")
lb_pais.place(width=60, height=20, x=210, y=210)
input_pais = Entry(root, font="Time 10")
input_pais.place(width=80, height=20, x=270, y=210)

lb_email = Label(root, text="Email", font="Time 10", anchor= "w")
lb_email.place(width=60, height=20, x=10, y=270)
input_email = Entry(root, font="Time 10")
input_email.place(width=250, height=20, x=100, y=270)
#------------------------------------------------------------------------------------------------------------------------------
# Funções

def adicionar():
    nome = input_nome.get()
    tel = input_tel.get()
    end = input_end.get()
    dist = input_dist.get()
    pais = input_pais.get()
    email = input_email.get()

    with open("agenda.txt", "a") as arquivo:
        arquivo.write(nome + "\n" + tel + "\n" + end + "\n" + dist + "\n" + pais + "\n" + email)

    messagebox.showinfo("Angenda", "Cadastro Efetuado com Sucesso!")
    
    input_nome.delete("0", "end")
    input_tel.delete("0", "end")
    input_end.delete("0", "end")
    input_dist.delete("0", "end")
    input_pais.delete("0", "end")
    input_email.delete("0", "end")

def procurar():
    nome = input_nome.get()

    with open("agenda.txt", "r") as arquivo:
        for linha in arquivo:
            if nome in linha:
                a_tel = (arquivo.readline())
                a_end = (arquivo.readline())
                a_dist = (arquivo.readline())
                a_pais = (arquivo.readline())
                a_email = (arquivo.readline())

                lb_nome_busca = Label(root, text=linha, font="Times 10", anchor="w")
                lb_nome_busca.place(width=250, height=30, x=20, y=360)
                lb_tel_busca = Label(root, text=a_tel, font="Times 10", anchor="w")
                lb_tel_busca.place(width=250, height=30, x=20, y=380)
                lb_end_busca = Label(root, text=a_end, font="Times 10", anchor="w")
                lb_end_busca.place(width=250, height=30, x=20, y=400)
                lb_dist_busca = Label(root, text=a_dist, font="Times 10", anchor="w")
                lb_dist_busca.place(width=250, height=30, x=20, y=420)
                lb_pais_busca = Label(root, text=a_pais, font="Times 10", anchor="w")
                lb_pais_busca.place(width=250, height=30, x=20, y=440)
                lb_email_busca = Label(root, text=a_email, font="Times 10", anchor="w")
                lb_email_busca.place(width=250, height=30, x=20, y=460)
            else:
                messagebox_showerror("Agenda", "Cadastro Não Encontrado!")
                break

#------------------------------------------------------------------------------------------------------------------------------
# Botões

b_adicionar = Button(root, text="Adicionar", command=adicionar, font="Time 10", bg=azul, fg=branco)
b_adicionar.place(width=80, height=30, x=70, y=310)

b_procurar = Button(root, text="Pesquisar", command=procurar, font="Time 10", bg=azul, fg=branco)
b_procurar.place(width=80, height=30, x=240, y=310)

#------------------------------------------------------------------------------------------------------------------------------
root.mainloop()
```
